```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('solana.csv')
data.head()
```

```python
data.describe()
```

```python
data['coin_name'].value_counts()
```

```python
X = data[['market_cap', 'total_volume']]
y = data['price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

This Projects explains the use of predictive machine learn algorithms. The dataset is the list of solana price from April 11, 2020 to February 31,2024
